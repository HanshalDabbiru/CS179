```python
def brute_force_marginal(model, query_var):
    """
    Brute-force marginal inference over a single query variable.
    Assumes model is already conditioned.
    """
    from itertools import product
    import numpy as np

    other_vars = [v for v in model.vars if v != query_var]
    marginal_vals = np.zeros(query_var.card)

    domains = [range(v.card) for v in other_vars]
    for assignment in product(*domains):
        assign_dict = {v: val for v, val in zip(other_vars, assignment)}
        for q_val in range(query_var.card):
            full_assign = assign_dict.copy()
            full_assign[query_var] = q_val
            marginal_vals[q_val] += model.evaluate(full_assign)

    marginal_vals /= marginal_vals.sum()
    return marginal_vals

def predict_mvp(df_new, model, discretizers, stat_vars, stat_gm_vars, mvp_var):
    import pandas as pd
    from copy import deepcopy

    # Discretize new stats
    for _, _, name in stat_vars:
        bins = discretizers[name]
        df_new[name + '_d'] = pd.cut(df_new[name], bins=bins, labels=False, include_lowest=True)

    predictions = []
    for _, row in df_new.iterrows():
        evidence = {
            var: int(row[name + '_d'])
            for (idx, _, name), var in zip(stat_vars, stat_gm_vars)
        }

        # Copy and condition model
        conditioned = deepcopy(model)
        conditioned.condition(evidence)

        # Brute-force marginal over MVP variable
        marginal = brute_force_marginal(conditioned, mvp_var)
        prob_mvp = marginal[1]  # Probability that MVP = 1
        predictions.append(prob_mvp)

    df_new['P_MVP'] = predictions
    predicted_mvp_row = df_new.loc[df_new['P_MVP'].idxmax()]
    return predicted_mvp_row
```

```python
discretizers = {
    'VORP': [0, 2, 4, 6, 8, 10],
    'WS': [0, 5, 10, 15],
    'OWS': [0, 3, 6, 9, 12],
    'FTA': [0, 100, 200, 300, 400, 500]
}

for col in ['VORP', 'WS', 'OWS', 'FTA']:
    df[col + '_d'] = pd.cut(df[col], bins=discretizers[col], labels=False, include_lowest=True)
    
stat_vars = [
    (0, len(discretizers['VORP']) - 1, 'VORP'),
    (1, len(discretizers['WS']) - 1, 'WS'),
    (2, len(discretizers['OWS']) - 1, 'OWS'),
    (3, len(discretizers['FTA']) - 1, 'FTA'),
]

stat_gm_vars = [gm.Var(idx, bins) for idx, bins, _ in stat_vars]
mvp_var = gm.Var(len(stat_vars), 2)

factors = []

# For each stat -> MVP dependency
for idx, bins, name in stat_vars:
    stat_var = gm.Var(idx, bins)
    counts = np.zeros((bins, 2))  # shape: (stat_bin, mvp)
    
    for _, row in df.iterrows():
        s_bin = row[name + '_d']
        if pd.isna(s_bin):
            continue
        s_bin = int(s_bin)
        mvp = int(row['Won_MVP'])
        counts[s_bin, mvp] += 1

    probs = (counts + 1) / (counts.sum(axis=1, keepdims=True) + 2)  # Laplace smoothing
    f = gm.Factor([stat_var, mvp_var])
    f.vals = probs
    factors.append(f)

model = gm.GraphModel()
model.addFactors(factors)

mvp_row = predict_mvp(df, model, discretizers, stat_vars, stat_gm_vars, mvp_var)
print(mvp_row)
print("Predicted MVP:", mvp_row['Player'])
```
